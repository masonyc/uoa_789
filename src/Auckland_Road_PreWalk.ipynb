```python
import pandas as pd

output_path = "."
LAYERS=5
```

```python
street_nodes_df = pd.read_csv(f"{output_path}/datasets/akl_street_nodes.csv")
street_edges_df = pd.read_csv(f"{output_path}/datasets/akl_street_edges.csv")

print(street_nodes_df.columns)
```

```python
prewalk_df = pd.DataFrame()
i=0
for index in street_nodes_df.index:
    row = street_nodes_df.iloc[[index]]
    current_layer_index = [index]
    new_node_df = pd.DataFrame(row.values,columns=street_nodes_df.columns)
    visited = [index]
    for layer in range(1,LAYERS+1):
        next_layer_index = []
        layer_restaurant_count,layer_amenity_count,layer_education_count,layer_healthcare_count,layer_shop_count,layer_cloth_count = 0,0,0,0,0,0

        layer_total_poi,layer_average_poi_distance = 0,0.0

        for cur_ind in current_layer_index:
            street_edges = street_edges_df[street_edges_df["source_street"]==cur_ind]
            street_edges = pd.concat([street_edges, street_edges_df[street_edges_df["target_street"]==cur_ind]])
            #print(f"Layer {layer} for start node index {index}, number of edges {len(street_edges)}")
            for edge_index,edge_row in street_edges.iterrows():
                edge_source,edge_target= edge_row["source_street"],edge_row["target_street"]

                target_node_index = None
                if edge_source == cur_ind:
                    target_node_index = int(edge_target)
                else:
                    target_node_index = int(edge_source)

                if target_node_index in next_layer_index or target_node_index in visited:
                    continue
                else:
                    next_layer_index.append(target_node_index)
                target_node = street_nodes_df.iloc[target_node_index]

                layer_restaurant_count += target_node["restaurant"]
                layer_amenity_count += target_node["amenity"]
                layer_education_count += target_node["education"]
                layer_healthcare_count += target_node["healthcare"]
                layer_shop_count += target_node["shop"]
                layer_cloth_count += target_node["cloth"]
                layer_average_poi_distance += target_node["Average_POI_Distance"]

        layer_poi_count = layer_restaurant_count + layer_amenity_count + layer_education_count + layer_healthcare_count+layer_shop_count+layer_cloth_count

        new_node_df[f"Layer_{layer}_raw_restaurant"] = layer_restaurant_count
        new_node_df[f"Layer_{layer}_raw_amenity"] = layer_amenity_count
        new_node_df[f"Layer_{layer}_raw_education"] = layer_education_count
        new_node_df[f"Layer_{layer}_raw_healthcare"] = layer_healthcare_count
        new_node_df[f"Layer_{layer}_raw_shop"] = layer_shop_count
        new_node_df[f"Layer_{layer}_raw_cloth"] = layer_cloth_count
        new_node_df[f"Layer_{layer}_raw_poi_count"] = layer_poi_count
        new_node_df[f"Layer_{layer}_raw_average_poi_distance"] = layer_average_poi_distance
        if layer == 1:
            new_node_df[f"Layer_{layer}_agg_restaurant"] = layer_restaurant_count
            new_node_df[f"Layer_{layer}_agg_amenity"] = layer_amenity_count
            new_node_df[f"Layer_{layer}_agg_education"] = layer_education_count
            new_node_df[f"Layer_{layer}_agg_healthcare"] = layer_healthcare_count
            new_node_df[f"Layer_{layer}_agg_shop"] = layer_shop_count
            new_node_df[f"Layer_{layer}_agg_cloth"] = layer_cloth_count
            new_node_df[f"Layer_{layer}_agg_poi_count"] = layer_poi_count
            new_node_df[f"Layer_{layer}_agg_average_poi_distance"] = layer_average_poi_distance / (layer_poi_count  if layer_poi_count > 0 else 1)
        else:
            new_node_df[f"Layer_{layer}_agg_restaurant"] = new_node_df[f"Layer_{layer-1}_agg_restaurant"] + layer_restaurant_count
            new_node_df[f"Layer_{layer}_agg_amenity"] = new_node_df[f"Layer_{layer-1}_agg_amenity"] + layer_amenity_count
            new_node_df[f"Layer_{layer}_agg_education"] = new_node_df[f"Layer_{layer-1}_agg_education"] + layer_education_count
            new_node_df[f"Layer_{layer}_agg_healthcare"] = new_node_df[f"Layer_{layer-1}_agg_healthcare"] + layer_healthcare_count
            new_node_df[f"Layer_{layer}_agg_shop"] = new_node_df[f"Layer_{layer-1}_agg_shop"] + layer_shop_count
            new_node_df[f"Layer_{layer}_agg_cloth"] = new_node_df[f"Layer_{layer-1}_agg_cloth"] + layer_cloth_count
            new_node_df[f"Layer_{layer}_agg_poi_count"] = new_node_df[f"Layer_{layer-1}_agg_poi_count"] + layer_poi_count
            new_node_df[f"Layer_{layer}_agg_average_poi_distance"] = new_node_df[f"Layer_{layer - 1}_agg_average_poi_distance"] + layer_average_poi_distance / (layer_poi_count  if layer_poi_count > 0 else 1)

        current_layer_index = next_layer_index

    prewalk_df=pd.concat([prewalk_df,new_node_df])

    i+=1
    if i % 2000 == 0:
        print(f"Finish {i}")
print("Done")
prewalk_df.to_csv(f"{output_path}/datasets/akl_prewalked_nodes.csv",index=False)
```
